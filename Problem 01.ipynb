###  Loading libraries

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd 
sns.set_style("darkgrid")
plt.rcParams['figure.figsize'] = [9, 6]
```

###  Make dataset

```python
CovMat = [[1,0],[0,1]]

Mean_negative = [-3,-3]
Mean_positive = [+3,+3]

class_negative = np.random.multivariate_normal(Mean_negative,CovMat,5000)
class_positive = np.random.multivariate_normal(Mean_positive,CovMat,5000)

class_negative = np.append(class_negative,[[-1]]*5000,axis = 1)
class_positive = np.append(class_positive,[[1]]*5000,axis = 1)

dataset = np.append(class_negative,class_positive,axis=0)
```

### Visualize

```python

dataframe = pd.DataFrame(data=dataset,columns=["Feature-1","Feature-2","Class"])
sns.scatterplot(data=dataframe,x="Feature-1",y="Feature-2",hue="Class")
```

###  Save dataset

```python
with open('dataset-linear.npy', 'wb') as f:
    np.save(f, dataset)
```
